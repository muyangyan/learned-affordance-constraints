```python
from models.ga import GAConstraintLearner, iterate_vars
from data.toy_dataset import ToyDataset
```

```python
def generate_library(bias_file, max_vars):
    vars = ['A', 'B', 'C', 'D', 'E', 'F']
    assert max_vars <= len(vars) 

    library_of_predicates = []

    with open(bias_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if line[0] == '%':
                continue
            pred_str, remainder = line.split('(')
            pred_name, remainder = remainder.split(', ')
            num_args = remainder.split(')')[0]
            if pred_str == 'body_pred':
                print(pred_name, num_args)
                library_of_predicates += iterate_vars(pred_name, vars[:max_vars], int(num_args))
    return library_of_predicates
```

```python
library_of_predicates = generate_library('prolog/general_bias.pl', 4)

background_knowledge = "prolog/bk.pl"
pos_examples = []
neg_examples = []


target_idx = 1
trainset = ToyDataset('data', mode='graph', split='train')
for idx, data in enumerate(trainset):
    if data.z[target_idx] == 1:
        pos_examples.append(f'x{idx}_0')
    else:
        neg_examples.append(f'x{idx}_0')
    


# Create and run GA
ga = GAConstraintLearner(
    library_of_predicates=library_of_predicates,
    background_knowledge=background_knowledge,
    pos_examples=pos_examples,
    neg_examples=neg_examples,
    population_size=5,
    generations=99,
    elite_fraction=0.1,
    crossover_rate=0.8,
    mutation_rate=0.2,
    alpha=2.0,
    beta=1.0
)
best_gene, best_fit = ga()  # Calls the GA loop
print(f"BEST GENE: {best_gene}, FITNESS: {best_fit:.3f}")
```
